```python
df_model = pd.read_csv('模型数据0311.csv')
```

```python
df_model.info()
```

```python
df_model.drop('passenger_id',axis=1,inplace=True)
```

```python
df_model[df_model['reg_days']==0]
```

```python
df_key = df_model.copy()
```

```python
import numpy as np
import xgboost as xgb
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_key.fillna(999, inplace=True)
```

```python
X = df_key.loc[:, df_key.columns != 'is_30d_unpay']
y = df_key['is_30d_unpay']
```

```python
X_train, X_test, y_train, y_test = \
        train_test_split(X, y, test_size=0.3, 
                         shuffle=True, random_state=123, stratify=y)
np.bincount(y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
predictions_tree = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions_tree))
```

```python
train_tree = dtree.predict(X_train)
```

```python
print(classification_report(y_train,train_tree))
```
